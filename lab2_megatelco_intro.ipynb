## Clean up the data
Did you notice anything unusual about the "house" amounts?

How about the handset price and income?

```python
df_clean = df[(df['house'] > 0) & (df['income'] > 0) & (df['handset_price'] < 1000)]
df_clean.describe()
```

# Add Derived Attribute

Add a column with the income in K (thousands)

```python
df_clean['income_k'] = (df_clean['income']/1000).round(2)
df_clean.head()
```

## Summarize the Data
 What is average house price at the two levels of leave?

 Conclusion?  Stayers have substantially more expensive homes.

```python
df_clean[['house','leave']].groupby('leave').mean().round(2)
```

## Additional Data Preparation

 Assignment Tasks
- Add at least two additional steps to clean or filter the dataset

```python
df_clean['over_15mins_calls_per_month'].fillna(0, inplace=True)
df_clean['over_15mins_calls_per_month'] = df_clean['over_15mins_calls_per_month'].astype(int)

college_mapping = {'one': 1, 'zero': 0}
df_clean['college'] = df_clean['college'].map(college_mapping).astype('int16')

df_clean.head()
```

## Data Visualization

 Assignment Tasks
- Add at least two charts to show a summary of any major trends

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='reported_satisfaction', data=df_clean, palette='viridis')
plt.title('Distribution of Reported Satisfaction')
plt.xlabel('Reported Satisfaction')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(12, 8))
sns.boxplot(x='reported_satisfaction', y='income', data=df_clean, palette='Set2')
plt.title('Boxplot: Distribution of Income by Reported Satisfaction')
plt.xlabel('Reported Satisfaction')
plt.ylabel('Income')
plt.show()
```
